```python
import spacy

TAG_MAP = {
 'NNS': {'pos': 'PLURAL_NOUN'},
}

def set_custom_attributes(doc):
    for token in doc:
        if token.tag_ in TAG_MAP:
            tag_info = TAG_MAP[token.tag_]
            for attr, value in tag_info.items():
                setattr(token._, attr, value)
    return doc

nlp = spacy.blank('en')
tagger = nlp.create_pipe('tagger')
for tag in TAG_MAP.keys():
    tagger.add_label(tag)
nlp.add_pipe(tagger, name="tagger")
nlp.add_pipe(set_custom_attributes, last=True)

# Test the custom attributes
text = "Apples are delicious fruits."
doc = nlp(text)
for token in doc:
    print(token.text, token.tag_, token._.pos)

# Output:
# Apples NNS PLURAL_NOUN
# are VBP
# delicious JJ
# fruits NNS PLURAL_NOUN
# . . .
```
